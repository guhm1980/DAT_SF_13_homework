```python
#numeric continuous variables: age; balance; day; duration; campaign; pdays; previous; 
#categorical variables:  job; marital; education; default; housing; loan; contact; month; poutcome
#target variable: y
```

```python
#QUESTION 2: Learn about label encoders at the following link and use what you learn to transform the features to numerical features.
```

```python
categorical = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']
encoder = {}
for x in categorical:
    numb = preprocessing.LabelEncoder()
    numb.fit(bank[x])
    encoder[x] = numb
    bank[x] = numb.transform(bank[x])
    
bank.head()
```

```python
#Question 3 - Build a decision tree model to predict whether a prospect will buy the product.
from sklearn.tree import DecisionTreeClassifier, ExtraTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
xColumns = ['job', 'marital', 'education', 'contact', 'month', 'poutcome', 'default', 'housing', 'loan']
X = bank[xColumns].values
y = bank['y'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print y_pred
print y_test
```

```python
from sklearn import metrics
import numpy as np
np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

```python
pd.DataFrame({'feature':xColumns, 'importance':clf.feature_importances_})
```

```python
#Question 4 - Evaluate the accuracy of your decision tree model using cross validation.
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import KFold
```

```python
scores = cross_val_score(clf, X, y, cv=KFold(len(X), n_folds=5,
                           shuffle=True, random_state=0))
scores
```

```python
#scores seem consistent around 86%
```
